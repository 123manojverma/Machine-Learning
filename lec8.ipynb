### continue, break

```python
# break statement
i=0
while(i<=5):
    print(i)
    i+=1
    if(i==3):
        break
```

```python
for i in range(10):
    print(i)
    if i==4:
        break
```

```python
# Generate random numbers between 1 to 10 till you get 5
import random
while True:
    #infinite loop
    r=random.randint(1,10)
    print(r)
    # r will be an random no. b/w 1 to 10
    # it will generate a random number between 1 to 10
    if(r==5):
        break
```

```python
# continue statement skips an iteration
# i want to skip all multiple of 3
for i in range(2,100,2):
    if i%3==0:
        continue
    print(i,end=" ")
```

### Linear Search

```python
a=[2,6,9,3,4,1,5]
b=9
# inbuilt function
print(a.index(b))
```

```python
# without the inbuilt function
# find the index of 'b' in 'a'
for i in range(len(a)):
    if a[i]==b:
        print(i)
        break
```
