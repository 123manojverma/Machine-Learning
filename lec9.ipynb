# Functions

A function is a block of code which only runs when it is called.

In Python a function is defined with the ‘def’ keyword

Functions can accept any number of arguments of any type.

Functions can optionally return data as well. This data may also be of any type. You can return more than one value as well.

Arguments can have default values as well.

While calling functions, the user may define arguments as keyword pairs.

```python
def tea():
    print("Boil water with tea leaves and sugar")
    print("Add Milk")
    print("Let it boil 4-5 times")

print("Not a part of the function")
def coffee():
    print("Boil some milk")
    print("Pour it in cup with some coffee")
    print("Stir")
```

```python
# calling tea function
tea()
for i in range(1):
    tea()
coffee()
```

## Map Function

```python
# Write a program to print prime no. in a range
def square(n):
    return n**2
a=[1,2,3,4,5]
b=list(map(square,a))
print(b)
```

```python
def prime(r):
    for i in range(2,r):
        if(r%i==0):
            return 
    return r

r=int(input("Enter the range: "))
ans=list(map(prime,range(2,r)))
print(ans)
```

## Lambda Functions

It's accept just one single statement

```python
def square(x):
    return x**2
```

```python
square2=lambda x:x**2
print(square2(850))
```

```python
# with multiple data input
multiply1=lambda a,b:a*b
def multiply(a,b):
    return a*b
```

```python
multiply1(19,56)
```

```python
# with multiple data return
d=lambda b:[b**2,b+b]
d(10)
```
