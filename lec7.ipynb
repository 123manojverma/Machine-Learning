For loop

It allows us to loop/Iterate over a range


```python
# Print the squares odd numbers from 1 to 10
for i in range(1,10,2):
    print(i**2,end=" ")
    # print(type(i))
```

```python
# Print the cube of all odd numbers from 10 to 1 (Reverse)
for i in range(9,0,-2):
    print(i**3,end=' ')
```

```python
n=int(input("Enter the number"))
for i in range(n,0,-1):
    for j in range(1,i+1):
        print(j,end=' ')
    print()
```

```python
# Spacing 
'''
0001000
0023200
0345430
4567654
'''
# We have n-i zeroes to print
# Increasing sequence -> start=i and end=2i-1
# Decreasing Sequence -> start 2i-2 and end=i
```

```python
n=int(input("Enter the number"))
for i in range(1,n+1):
    print("0"*(n-i),end="")
    for j in range(i,2*i):
        print(j,end="")
    for j in range((2*i)-2,i-1,-1):
        print(j,end="")
    print("0"*(n-i),end="")
    print()
```
