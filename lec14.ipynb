### Combinations

math module method -
math.comb(n,r)

- The no. of ways to choose r objects out of n total objects
- nCr = C(n,r) = nCr = n!/r!(n-r)!

````python
import math
math.comb(10,4) 
````

````python
def combination(n,r):
    num=math.factorial(n)
    den=math.factorial(r)*math.factorial(n-r);
    return num/den
n=int(input("Enter n: "))
r=int(input("Enter r: "))
print(combination(n,r))
````

### Permutations

math module method - math.perm(n,r)

````python
math.perm(10,4)
````

````python
def permutation(n,r):
    return math.factorial(n)/math.factorial(n-r)
n=int(input("Enter n: "))
r=int(input("Enter r: "))
print(permutation(n,r))
````

````python
def decimalToBinary(n):
    return bin(n).replace("0b","")
print(bin(20))
print(decimalToBinary(20))
````

````python
def decimalToBinary(n):
    l=[]
    while n>0:
        remainder=n%2
        l.append(str(remainder))
        n=n//2
    # Reverse the list
    l=list(reversed(l))
    return "".join(l)
print(decimalToBinary(20))
````

````python
# x is a binary string
def BinaryToDecimal(x):
    return int(x,2)
print(BinaryToDecimal("100"))
````

````python
def BinaryToDecimal(s):
    i=0
    result=0
    for digit in s[::-1]:
        result+=int(digit)*(2**i)
        i+=1
    return result
BinaryToDecimal("10100")
````

### Trailing zeroes in factorial 

- Given an integer A, return the number of trailing zeroes in A! i.e. factorial of A.

- Note: Your solution should run in logarithmic time complexity.

- Input : 5
- Output : 1

````python
def trailingZeroes(A):
    result=0
    den=5
    x=A//den
    while x>=1:
        result+=x
        den*=5
        x=A//den
    return result
trailingZeroes(21)
````
