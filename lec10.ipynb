# List Slicing

a[start:end:increment] in list

```python
a=[2,4,9,3,1,6]
print(a[1:5:2])
print(a[::-1])   # It will print the reverse list
print(a[4:1:-1])
```

# List Comprehension

```python
# Generate a list of squares of all numbers 1 to 100
a=[x**2 for x in range(1,101)]
print(a)
```

# Using map function

```python
def square(x):
    return x**2
res=list(map(square,range(1,101)))
print(res)
```

```python
a=[27,24,21,46,48,6,25,38,10,40]
b=[]
# Using for loop
for x in a:
    if x%2==0:
        b.append(x)
print(b)
```

```python
# Using list comprehension
b=[x for x in a if(x%2==0)]
print(b)
```

# Tuples

- Tuples are used to store multiple items in a single variable.

- A tuple is a collection which is ordered and immutable (unchangeable).

- Tuples are written with round brackets.

- Tuple items are indexed, the first item has index [0], the second item has index [1] etc.

- Tuples can be created with tuple constructor - tuple()

- Tuples can also be created by simply writing multiple values with comma separation.

- Tuples are processed faster than lists in Python.

```python
# Read only list
continents=('Asia','Africa','North America','South America','Europe')
print(type(continents))
```

```python
# continents[0]='India' 'tuple' object does not support item assignment
print(continents[0])
```

# Write a function which returns the product of all its arguments.

```python
def product(*nums):
    print(nums,type(nums))
product(10,20,30,40,50,60,4,2,5,12)
```

```python
def speak(*words):
    print(words,type(words))
speak("Hello","my","name","is","Manoj")
```

```python
def product(*args):
    p=1
    for number in args:
        p=p*number
    return p
product(17,28,31)
```

```python
# * - spread operator
a=[2,5,6,1,8]
print(*a)
```
