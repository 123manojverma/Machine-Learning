```python
def person(name,age,profession, **kwargs):  # that will not handle extra arguments we pass that's why we use kwargs
    print(name,age,profession,kwargs)
    print(type(kwargs))
```

```python
person('Peter Parker',20,'Photographer',city='New York',villains=['Venom','Green Goblin'])
```

```python
def hogwarts(house1,house2,house3,house4,*args,**kwargs):
    print('House are',house1,house2,house3,house4)
    print()
    print('Args :',args)
    print()
    print('Kwargs :',kwargs)
```

```python
hogwarts(
    'Grayfinder',
    'Racenclow',
    'Hufflepuf',
    'Slytherin',
    'mannat',
    'parampara',
    'Romus Lupin',
    headmaster='Albus Dumbledore',
    deputy_headmistress='Menerva MCGonagull'
)
```
