## Logarithm

The exponent or power to which a base must be raised to yield a given number. Expressed mathematically, x is the logarithm of n to the base b if bx=n, in which case one writes x =logbn. For example, 23=8; therefore, 3 is the logarithm of 8 to base 2, or 3 = log2 8.

Python functions for computing log;

- math.log(x)  -> returns log(x) with base=e
- math.log(x,a) -> returns log(x) with base=a
- math.log2(x)  -> returns log(x) with base=2
- math.log10(x)  -> returns log(x) with base=10

### GCD 
math module method - math.gcd(*integers)

```python
math.gcd(30,50)
```

```python
a=int(input())
b=int(input())

def gcd(a,b):
    while a!=0:
        a,b=b%a,a
    return b
print("GCD is",gcd(a,b))
```

### Prime Number - sqrt(N) method

```python
def isPrime(n):
    """
    This function returns True or False based on primality check of n.
    """
    
    # Search for factors in the range [2,sqrt(n)]
    # sq=n**(1/2)
    if n<=1:
        return False
    sq=int(math.sqrt(n))
    
    for i in range(2,sq+1):
        if n%i==0:
            return False
    return True
```

```python
print(isPrime(17))
isPrime(81)
```

```python
n=int(input())
for i in range(1,n+1):
    print(isPrime(i))
```
